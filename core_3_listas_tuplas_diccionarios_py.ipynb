```python
#Crea una lista de diccionarios llamada ventas, donde cada diccionario represente una venta

ventas = [
    {"fecha": "24-03-2025", "producto": "sandia", "cantidad": 2, "precio": 1500},
    {"fecha": "27-08-2023", "producto": "jabon", "cantidad": 15, "precio": 200},
    {"fecha": "24-03-2025", "producto": "carton de jugo", "cantidad": 6, "precio": 600},
    {"fecha": "24-03-2025", "producto": "kilo de carne", "cantidad": 1, "precio": 3500}
]

#Utiliza un bucle para iterar sobre la lista ventas y calcular los ingresos totales generados por todas las ventas
ingresos_totales = 0

for venta in ventas:
    ingresos_totales += venta["cantidad"] * venta["precio"]

print(ingresos_totales)

#Crea un diccionario llamado ventas_por_producto donde las claves sean los nombres de los productos
#y los valores sean la cantidad total vendida de cada producto.

ventas_por_producto = {
    "sandia": 2,
    "jabon": 15,
    "carton de jugo": 6,
    "kilo de carne": 1
}

#Utiliza este diccionario para identificar el producto que tuvo la mayor cantidad total vendida.
producto_mas_vendido = max(ventas_por_producto, key=ventas_por_producto.get)

print(producto_mas_vendido)

#Crea un diccionario llamado precios_por_producto donde las claves sean los nombres de los productos y los valores sean tuplas
#Cada tupla debe contener la suma de los precios de venta de todas las unidades vendidas y la cantidad total de unidades vendidas.

precios_por_producto = {}

for venta in ventas:
    producto = venta["producto"]
    ingresos = venta["cantidad"] * venta["precio"]
    cantidad = venta["cantidad"]

    if producto not in precios_por_producto:
        precios_por_producto[producto] = (ingresos, cantidad)
    else:
        suma_ingresos, cantidad_total = precios_por_producto[producto]

        precios_por_producto[producto] = (
            suma_ingresos + ingresos,
            cantidad_total + cantidad
        )

#Crea un diccionario llamado ingresos_por_dia donde las claves sean las fechas y los valores sean los ingresos totales generados en cada día.
#Utiliza un bucle para calcular los ingresos totales por día y almacenar estos valores en el diccionario.

ingresos_por_dia = {}

for venta in ventas:
    fecha = venta["fecha"]
    ingresos = venta["cantidad"] * venta["precio"]

    if fecha not in ingresos_por_dia:
        ingresos_por_dia[fecha] = ingresos
    else:
        ingresos_por_dia[fecha] += ingresos

print(ingresos_por_dia)


# Representación de Datos:
# Crea un diccionario llamado resumen_ventas donde las claves sean los nombres
# de los productos y los valores sean diccionarios anidados.

resumen_ventas = {}

for venta in ventas:
    producto = venta["producto"]
    cantidad = venta["cantidad"]
    ingresos = cantidad * venta["precio"]

    if producto not in resumen_ventas:
        resumen_ventas[producto] = {
            "cantidad_total": cantidad,
            "ingresos_totales": ingresos
        }
    else:
        resumen_ventas[producto]["cantidad_total"] += cantidad
        resumen_ventas[producto]["ingresos_totales"] += ingresos

# Calcular precio promedio
for producto in resumen_ventas:
    cantidad_total = resumen_ventas[producto]["cantidad_total"]
    ingresos_totales = resumen_ventas[producto]["ingresos_totales"]

    resumen_ventas[producto]["precio_promedio"] = (
        ingresos_totales / cantidad_total
    )

print(resumen_ventas)
```
